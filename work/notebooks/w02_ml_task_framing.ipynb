## 2. Target or proxy

## Target or Proxy

The target (proxy) is the Click-Through Rate (CTR) of the content. CTR is an observed performance metric that indicates how often users click on the content after seeing it. The model will use features such as search volume, competition, CPC, content type, search intent, and word count to estimate CTR.

```python
target = "ctr"

print("Target Column:", target)

if target in df.columns:
    print("\nSample Target Values:")
    print(df[target].head())
else:
    print("Target column not found.")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The success metric for this task is **Mean Absolute Error (MAE)**. MAE measures the average difference between the predicted CTR and the actual CTR values. A lower MAE indicates that the model predicts content performance more accurately, making it useful for prioritizing content refresh decisions.

```python
success_metric = "Mean Absolute Error (MAE)"

print("Success Metric:", success_metric)
print("Goal: Lower MAE means better prediction accuracy.")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule cannot accurately predict content performance because many factors influence the outcome, such as search volume, competition, CPC, content type, search intent, and content length. The relationship between these features is complex and changes over time. A machine learning model can learn these patterns from historical data and provide better predictions than manually written rules. This helps in making better decisions about which content should be refreshed or prioritized.
